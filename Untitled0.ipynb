```python
# =======================================================
# [논리 퍼즐 1] 양수, 음수, 0 판단 코드
# =======================================================

# 1. 판별하고 싶은 숫자를 여기에 입력해 주세요.
number = 8

# 2. 논리 시작: 만약 (if) 숫자가 0보다 크면 양수
if number > 0:
    print(f"입력된 숫자 {number}는 양수입니다.")

# 3. 그렇지 않고 만약 (elif) 숫자가 0보다 작으면 음수
elif number < 0:
    print(f"입력된 숫자 {number}는 음수입니다.")

# 4. 위의 모든 조건이 아니라면 (else) 0
else:
    print(f"입력된 숫자 {number}는 0입니다.")
```
